Grafica de la clase (Minimos cuadrados)

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
x = np.array([-3, -1, 2])
y = np.array([4.25, 4.75, 5.5])

# Definición del polinomio de grado 1: y = mx + b
m = 1 / 4
b = 5

def linear_function(x):
    return m * x + b

# Generar valores para la gráfica
x_vals = np.linspace(min(x) - 1, max(x) + 1, 500)
y_vals = linear_function(x_vals)

# Graficar
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', label='Puntos dados')
plt.plot(x_vals, y_vals, label='Recta: y = mx + b', color='blue')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('Representación de la Recta: y = mx + b')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()

```
