### Aplicacion de polinomios de Taylor 
a. $$\frac{1}{25 \cdot x^2 + 1} , \quad x_0 = 0$$ 

```python
import sympy as sp

##Definimos la funcion que nos ayudara calcular el polinomio de taylor 
def poli_taylor(f,a,n):
    #Variable simbolica 
    x = sp.symbols('x')
    
    #Inicializacion del polinomio 
    taylor=0

    for i in range(n + 1):
        derivada = sp.diff(f, x, i)  # Derivada i-ésima
        termino = (derivada.subs(x, a) / sp.factorial(i)) * (x - a)**i
        taylor += termino
    
    return sp.simplify(taylor)
```

Una vez definida la funcion que nos ayuda a calcular el polinomio de taylor de orden n de la funcion f(x) en el punto, procedemos a realizar la funcion que nos ayude a graficar: 

```python
import matplotlib.pyplot as plt 
import numpy as np

def graficar(f,a,n): #comprobacion de la grafica.
    # Definir la variable simbólica y calcular el polinomio de Taylor
    x = sp.Symbol('x')
    taylor1 = poli_taylor(f, a, n)

    # Convertir funciones simbólicas a funciones numéricas
    f_num = sp.lambdify(x, f, 'numpy')        # Función original
    xs=np.linspace(-2, 2, 500)
    ys=f_num(xs)

    taylor_num1 = sp.lambdify(x, taylor1, 'numpy')  # Polinomio de Taylor
    taylor_vals=taylor_num1(xs)


    plt.figure(figsize=(8, 6))
    plt.plot(xs, ys, label="Función Original", color="blue")
    plt.plot(xs, taylor_vals, label=f"Polinomio de Taylor (grado {n})", color="red", linestyle="--")
    plt.title("Función y Polinomio de Taylor")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.xlim(-1,2)
    plt.ylim(-3, 2)
    plt.grid()
    plt.show()
```

### Aplicacion de Lagrange

a.$$\frac{1}{25 \cdot x^2 + 1} , \quad x_0 = 0$$ 

Para la aplicacion le Lagrange necesitamos defnir puntos a dependiendo del grado n de f(x) en este caso es de grado 2 por lo que se tendran n+1 puntos en este caso el numero de puntos es 3. Por lo que tenemos: 

p1= (-0.2 ; 0.5), p2=(0 ; 1), p3=(0.2 ; 0.5), Estos puntos son tomados de la siguiente manera: p1 es el punto de inicio, p2 es el punto medio y p3 el punto del fin en base a la grafica del metodo de taylor para mayor facilidad y versatilidad. 

```python
##Importamos la libreria que nos ayudara a obtener el polinomio de lagrange
from scipy.interpolate import lagrange

x=[-0.2, 0, 0.2]
y=[0.5, 1, 0.5]

plagrange=lagrange(x,y)
print("El polinomio de lagrange es:\n " +str(plagrange))

```

```python
import matplotlib.pyplot as plt 
import numpy as np


# Rango de valores de x
x = np.linspace(-2, 2, 500)

# Definición de la función original y el polinomio de interpolación
func = 1 / (25 * (x**2) + 1)
polilagrange = -12.5 * x**2 + 1

# Crear la figura y graficar
plt.figure(figsize=(10, 6))  # Configuración del tamaño de la figura
plt.plot(x, func, label='Función original', color='blue')
plt.plot(x, polilagrange, label='Polinomio de interpolación', linestyle='--', color='red')

#Definciin de los puntos para la grafica de interpolacion 
xs=[-0.2, 0, 0.2]
ys=[0.5, 1, 0.5]
plt.scatter(xs, ys, label= 'Puntos', color='blue', zorder=5)

# Configuración de la gráfica
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interpolación con el polinomio de Lagrange')
plt.legend()
plt.grid(True)
plt.xlim(-0.4, 0.4)
plt.ylim(-0.25, 1.25)

# Mostrar la gráfica
plt.show()
```
